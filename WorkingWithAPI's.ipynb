```python
import requests
import pandas as pd

df = pd.DataFrame(columns=['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count'])

# Your API key
api_key = '8265bd1679663a7ea12ac168da84d2e8'

for i in range(1, 429):
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key={api_key}&language=en-US&page={i}')
    if response.status_code == 200:
        results = response.json().get('results', [])

        temp_df = pd.DataFrame(results)[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]

        df = pd.concat([df, temp_df], ignore_index=True)
    else:
        print(f"Failed to retrieve data for page {i}. Status code: {response.status_code}")

df.to_csv('top_rated_movies.csv', index=False)
```
