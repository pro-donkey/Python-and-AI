```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

##### In this scenario, product sales are the category data that is visually represented by a 2-D bar chart. Product categories (A, B, C, etc.) are shown on the X-axis, while matching sales numbers are displayed on the Y-axis. Each category is represented by a bar, the length of which reflects the sales volume. Exact values are given via the data labels above each bar, which facilitates comprehension. Context is provided via axis labels and titles, like "Product Sales Summary." Because it makes comparing sales across items easier overall, this chart is frequently used in data analysis and business.

```python


categories = ['a', 'b', 'c', 'd', 'e']
sales = [120, 90, 150, 80, 110]
data = {"Product":categories,
        "Sales":sales}
df = pd.DataFrame(data)
plt.bar(categories, sales, color='skyblue')

for i in range(len(categories)):
    plt.text(i, sales[i] + 5, str(sales[i]), ha='center')

plt.xlabel('Product Categories')
plt.ylabel('Sales Volume')
plt.title('Product Sales Summary')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.show()
```

##### Bar charts with clusters are useful for comparing various categories within groups. Age groups are divided in this scenario, and preferences for the action, comedy, and drama genres are shown for each group. This kind of data organisation makes it easier for viewers to identify patterns and preferences across various age groups. With the help of this visualisation, the entertainment business may make more informed judgements by better grasping the subtleties of movie genre preferences across different age groups.

```python
data = {
    "Age Group": ["18-25", "26-35", "36-45"],
    "Action": [30, 20, 15],
    "Drama": [40, 50, 30],
    "Comedy": [20, 25, 35]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))

bar_width = 0.2

index = df.index

plt.bar(index - bar_width, df['Action'], width=bar_width, label='Action')
plt.bar(index, df['Drama'], width=bar_width, label='Drama')
plt.bar(index + bar_width, df['Comedy'], width=bar_width, label='Comedy')

plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Genre Preferences by Age Group')
plt.xticks(index, df['Age Group'])
plt.legend()


plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.show()
```

Plotting dots and connected dots are useful tools for showing patterns and
variations over time or under various circumstances. The three semesters' worth
of student scores are compared across each row in this dataset.
Every student's score is shown as a dot for each semester along a common axis in
a dot plot. This makes it simple to compare a student's performance over several
semesters.
A connected dot plot more clearly illustrates the trend over time by joining the
dots, which represent each student's results over semesters, with lines.
These graphic aids make it easier to spot trends in each student's growth or
decrease and make comparisons between their academic achievement across
several semesters.

```python
names = ["S" + str(i) for i in range(1, 6)]
data = {
    "Students": names,
    "Sem1": [75, 85, 70, 88, 90],
    "Sem2": [80, 88, 75, 90, 85],
    "Sem3": [90, 92, 85, 94, 88]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))

for i, student in enumerate(df["Students"]):
    plt.scatter(["Sem1", "Sem2", "Sem3"], df.loc[i, ["Sem1", "Sem2", "Sem3"]], label=student)

plt.xlabel("Semesters", fontsize=14, fontweight="bold")
plt.ylabel("Scores", fontsize=14, fontweight="bold")
plt.title("Student Scores Across Semesters (Dot Plot)", fontsize=16, fontweight="bold")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title="Students", fontsize=10, title_fontsize="12")

plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))

for i, student in enumerate(df["Students"]):
    plt.plot(["Sem1", "Sem2", "Sem3"], df.loc[i, ["Sem1", "Sem2", "Sem3"]], marker='o', label=student)


plt.xlabel("Semesters", fontsize=14, fontweight="bold")
plt.ylabel("Scores", fontsize=14, fontweight="bold")
plt.title("Student Scores Across Semesters (Connected Dot Plot)", fontsize=16, fontweight="bold")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title="Students", fontsize=10, title_fontsize="12")

plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

Pictograms are a type of visual data representation that use symbols or icons to
express information. Every row in this dataset denotes a student and the quantity
of books they have perused.
Making a pictogram:
1. Select an icon or symbol to stand in for each data unit (e.g., one book).
2. Establish the pictogram's scale, where each symbol (e.g., each book icon
indicates one read book) corresponds to a specific quantity of the data.
3. Using the data value for each student, draw the right amount of symbols.

```python
names = ["S" + str(i) for i in range(1, 6)]
data = {
    "Students": names,
    "Books Read": [3, 2, 4, 3, 2]
}

df = pd.DataFrame(data)

color_map = {1: 'skyblue', 2: 'salmon', 3: 'lightgreen', 4: 'gold', 5: 'orchid'}

plt.figure(figsize=(8, 5))

for i, (student, books) in enumerate(zip(df["Students"], df["Books Read"])):
    plt.barh(y=i, width=books, color=color_map[books])

plt.xlabel("Books Read", fontsize=14, fontweight="bold")
plt.ylabel("Students", fontsize=14, fontweight="bold")
plt.title("Pictogram: Books Read by Students", fontsize=16, fontweight="bold")
plt.yticks(range(len(df["Students"])), df["Students"], fontsize=12)
plt.xticks(fontsize=12)

plt.tight_layout()
plt.show()
```

Bubble charts are a useful way to visualize data points in three dimensions: the
x-axis, the y-axis, and the size of the bubbles. In your dataset, you have students'
scores in three subjects (Math, English, Science) as well as their extracurricular
hours.  
Each student would be represented by a bubble on the chart, with its position
determined by their Math and English scores, the size of the bubble indicating
their Science score, and the color (if used) indicating their extracurricular hours.
Creating a bubble chart in Excel:
1. Select your data range, including headers.
2. Go to the "Insert" tab on the Excel ribbon.
3. In the Charts group, click on "Insert Scatter (X, Y) or Bubble Chart."
4. Choose the bubble chart type you prefer.
5. Excel will generate the chart, and you can further customize it by adding
axis titles, changing bubble colors, adjusting bubble sizes, and so on.

```python

names = ["S" + str(i) for i in range(1, 6)]
data = {
    "Students": names,
    "Maths": [90, 88, 75, 92, 80],
    "English": [85, 90, 80, 85, 88],
    "Science": [92, 87, 78, 90, 85],
    "Extracurricular Activities": [15, 20, 10, 18, 12]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))

for i, student in enumerate(df["Students"]):
    plt.scatter(df.loc[i, "Maths"], df.loc[i, "English"], s=df.loc[i, "Science"] * 10,
                c=df.loc[i, "Extracurricular Activities"], cmap='plasma', alpha=0.7,
                label=student)

plt.xlabel("Maths Score", fontsize=14, fontweight="bold")
plt.ylabel("English Score", fontsize=14, fontweight="bold")
plt.title("Bubble Chart: Student Performance", fontsize=16, fontweight="bold")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

cbar = plt.colorbar()
cbar.set_label('Extracurricular Activities', fontsize=12)

plt.legend(fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### A Radar chart is a graphical method of displaying multivariate data in a twodimensional form. It uses a circular layout, with each variable represented by a spoke radiating from the center of the chart. In this context, the dataset includes students' scores in Math, English, and Science, which are plotted on separate axes. Connecting these points forms a polygon that illustrates the student's performance across these subjects. The distance from the center of the chart to each point on the polygon indicates the score for that subject. Additionally, extracurricular hours can be represented using different colors or labels on the radar plot, offering further insights into students' overall engagement and performance.

```python
names = ["S" + str(i) for i in range(1, 6)]
data = {
    "Students": names,
    "Maths": [90, 88, 75, 92, 80],
    "English": [85, 90, 80, 85, 88],
    "Science": [92, 87, 78, 90, 85],
    "Extracurricular Activities": [15, 20, 10, 18, 12]
}

df = pd.DataFrame(data)
categories = list(df.columns[1:])
num_vars = len(categories)

plt.figure(figsize=(10, 10))
plt.subplots_adjust(hspace=0.5, wspace=0.5)

for i, student in enumerate(df["Students"]):
    values = df.loc[i, categories].values

    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    values = np.concatenate((values, [values[0]]))
    angles += angles[:1]

    ax = plt.subplot(3, 2, i+1, polar=True)
    ax.fill(angles, values, color='skyblue', alpha=0.3)
    ax.plot(angles, values, color='skyblue', linewidth=2)
    ax.set_yticklabels([])
    plt.title(student)


plt.subplot(3, 2, 1)
plt.figlegend(loc='upper center', bbox_to_anchor=(0.5, 1.05), title="Subjects", title_fontsize='large', fontsize='large')
plt.suptitle("Radar Chart: Student Performance", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.show()
```

A polar chart, also known as a radar chart, visually represents multivariate data
in a circular manner, with each variable plotted along a spoke radiating from the
center of the chart. In this dataset, each student's scores in Math, English, and
Science are represented by points along the respective spokes, forming a polygon
that illustrates their performance profile across these subjects. The distance from
the center to each point on the polygon indicates the score for that subject,
allowing for easy comparison of performance across multiple dimensions for
each student.

```python

names = ["S" + str(i) for i in range(1, 6)]
data = {
    "Students": names,
    "Maths": [90, 88, 75, 92, 80],
    "English": [85, 90, 80, 85, 88],
    "Science": [92, 87, 78, 90, 85],
    "Extracurricular Activities": [15, 20, 10, 18, 12]
}


df = pd.DataFrame(data)
categories = list(df.columns[1:])
num_vars = len(categories)

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

angles += angles[:1]

plt.figure(figsize=(10, 8))

for i, student in enumerate(df["Students"]):
    values = df.loc[i, categories].values

    values = np.concatenate((values, [values[0]]))


    ax = plt.subplot(3, 2, i+1, polar=True)
    ax.fill(angles, values, color='skyblue', alpha=0.3)
    ax.plot(angles, values, color='skyblue', linewidth=2)
    ax.set_yticklabels([])
    plt.title(student)

plt.subplots_adjust(top=0.85)
plt.figlegend(categories, loc='upper center', bbox_to_anchor=(0.5, 0.85), title="Subjects", title_fontsize='large', fontsize='large')


plt.suptitle("Polar Chart: Student Performance", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.show()
```

A range chart visually represents the range of values for each category across
different data points or time periods. In this case, the chart displays the
temperature range for different cities throughout the week. - Each row represents a city (A, B, C, D, E). - The columns (Mon, Tue, Wed, Thu, Fri) represent different days of the week. - The numbers in each cell represent the temperature range for a specific city on
a specific day.
Viewers can quickly observe the temperature variations for each city and identify
trends or patterns in temperature changes throughout the week. Range charts are
effective for showing the spread or variability in data across multiple categories
or time points.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "City": ["A", "B", "C", "D", "E"],
    "Monday": [22, 18, 23, 20, 25],
    "Tuesday": [24, 21, 25, 23, 20],
    "Wednesday": [20, 19, 21, 19, 26],
    "Thursday": [25, 22, 24, 24, 23],
    "Friday": [23, 20, 22, 21, 20]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))

for i, city in enumerate(df["City"]):
    plt.plot(df.columns[1:], df.iloc[i, 1:], marker='o', label=city)

plt.title("Temperature Range for Different Cities Throughout the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Temperature")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "City": ["A", "B", "C", "D", "E"],
    "Monday": [22, 18, 23, 20, 25],
    "Tuesday": [24, 21, 25, 23, 20],
    "Wednesday": [20, 19, 21, 19, 26],
    "Thursday": [25, 22, 24, 24, 23],
    "Friday": [23, 20, 22, 21, 20]
}

df = pd.DataFrame(data)


plt.figure(figsize=(10, 6))

for i, city in enumerate(df["City"]):
    plt.plot(df.columns[1:], df.iloc[i, 1:], marker='o', label=city)

plt.title("Temperature Range for Different Cities Throughout the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Temperature")
plt.legend()
plt.grid(True)
plt.show()


plt.figure(figsize=(10, 6))

city_names = df["City"]
ranges = df.iloc[:, 1:].max(axis=1) - df.iloc[:, 1:].min(axis=1)

plt.barh(city_names, ranges, color='skyblue')
plt.title("Temperature Range Across Cities")
plt.xlabel("Temperature Range")
plt.ylabel("City")
plt.grid(axis='x')
plt.show()
```

Box-and-whisker plots provide a visual summary of the distribution of data,
including the median, quartiles, and potential outliers. In this dataset, two box
and-whisker plots are created for Math and English exam scores.  
Each plot consists of: - A box, representing the interquartile range (IQR) of the data. The bottom and
top of the box mark the first (Q1) and third (Q3) quartiles, respectively, while the
line inside the box represents the median. - Whiskers extending from the box, indicating the range of the data. They
typically extend 1.5 times the IQR from the quartiles, with outliers represented as
individual points beyond the whiskers.
For the Math exam, the scores range from 60 to 90, with the median around 80.
The English exam scores range from 70 to 92, with a median closer to 85. These
plots offer insights into the spread and central tendency of the exam scores for
each subject.

```python
data = {
    "Maths": [60, 70, 80, 90],
    "English": [70, 80, 90, 75]
}

plt.boxplot(data.values(), vert=True, patch_artist=True, notch=True, medianprops={'linewidth': 5})

plt.xticks([1, 2], data.keys())

plt.title('Box-and-Whisker Plots for Exam Scores')
plt.ylabel('Scores')
plt.show()
```

Univariate scatter plots depict the relationship between two variables by
plotting each data point along a single axis. In this dataset, hours of study are
plotted against corresponding scores. Each point represents a unique combination
of study hours and scores.  
For example, a point at (5, 70) indicates a score of 70 achieved after studying for
5 hours. Similarly, (8, 85) represents a score of 85 after 8 hours of study.
These plots help visualize the distribution of scores relative to study hours,
identifying potential trends or patterns between the two variables. They are useful
for exploring the relationship between variables and identifying any outliers or
clusters in the data.

```python
data = {
    "Hours": [5, 8, 10, 6, 12],
    "Scores": [70, 85, 92, 78, 94]
}

plt.scatter(data["Hours"], data["Scores"], color='blue', label='Data Points')
plt.grid(True, which='both', color='gray', linestyle='--')

plt.xlabel('Hours')
plt.ylabel('Scores')

z = np.polyfit(data["Hours"], data["Scores"], 1)
p = np.poly1d(z)
plt.plot(data["Hours"], p(data["Hours"]), color='red', label='Trend Line')


plt.legend()

plt.title('Scatter Plot of Scores vs Study Hours')
plt.show()
```
